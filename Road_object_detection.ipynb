```python
from ultralytics import YOLO
import torch
from torch.optim.lr_scheduler import CosineAnnealingLR

# Load YOLOv8 pre-trained model
model = YOLO('yolov8n.pt')  


# Define the configuration for training
model.train(
    data='/kaggle/working/sign-detection-1/data.yaml',  # Path to your dataset.yaml file
    imgsz=640,  # Image size
    batch=16,  # Batch size (adjust based on your GPU capacity)
    epochs=100,  # Number of epochs (you can adjust this as needed)
    lr0=0.01,  # Initial learning rate
    lrf=0.2,  # Learning rate final (multiplier for lr0 at the end of training)
    momentum=0.937,  # Momentum for the optimizer
    weight_decay=0.0005,  # Regularization (weight decay)  
    save_period=10,  # Save model checkpoints every 10 epochs
)
```

```python
Evaluate the Model After Training

results = model.val(data='/kaggle/working/sign-detection-1/data.yaml')  # Path to your test dataset.yaml

# Class-specific evaluation
class_results = results.metrics['class']
for class_name, (precision, recall, mAP50, mAP50_95) in class_results.items():
    print(f"Class: {class_name}")
    print(f"  Precision: {precision}")
    print(f"  Recall: {recall}")
    print(f"  mAP50: {mAP50}")
    print(f"  mAP50-95: {mAP50_95}")
```

```python
metrics = model.val()
print(metrics)
```
